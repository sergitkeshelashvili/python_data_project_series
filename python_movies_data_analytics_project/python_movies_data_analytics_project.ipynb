```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px

```

```python
df = pd.read_csv('movies_raw_data.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns = ['id', 'imdb_id', 'homepage', 'cast', 'tagline', 'overview', 'budget_adj'])
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna(how = 'any', subset = ['genres', 'director'])
```

```python
df['production_companies'] = df['production_companies'].fillna(0)
df['keywords'] = df['keywords'].fillna(0)
```

```python
df.isnull().sum()
```

```python
df['popularity'] = df['popularity'].round(2)
```

```python
df.insert(3, 'profit', df.revenue - df.budget)
```

```python
df.insert(4, 'roi', df.profit / df.budget)
```

```python
df['roi'] = df['roi'].round(2)
df.head()
```

```python
df.roi.value_counts()
```

```python
non_finite_values = ~np.isfinite(df['roi'])
```

```python
non_finite_values.sum()
```

```python
df['roi'] = df['roi'].replace([np.inf, -np.inf], np.nan)
```

```python
df1 = df [['popularity', 'budget', 'revenue', 'profit', 'roi', 'vote_count', 'release_year']]
```

```python
print(df1.isin([np.inf, -np.inf]).any())  # Check for infinite values
print(df1.isna().any())                   # Check for NaN values
```

```python
numeric_cols = df1.select_dtypes(include=np.number).columns
print("Columns with infinite values:")
print(df1[numeric_cols].isin([np.inf, -np.inf]).any())
```

```python
print("Columns with NaN values:")
print(df1[numeric_cols].isna().any())
```

```python
df1_clean = df1[numeric_cols].replace([np.inf, -np.inf], np.nan).dropna()
```

```python
df1_clean = df1[numeric_cols].replace([np.inf, -np.inf], np.nan)
df1_clean = df1_clean.fillna(df1_clean.mean())  # Or use a specific value, e.g., 0
```

```python
df1_clean.hist(bins=20, figsize=(14,12))
plt.show()
```

```python
df2 = df.groupby('release_year')['roi'].mean()
df2.plot(kind = 'line') 
```

```python
df3 = df.groupby('release_year')['popularity'].sum().plot(kind = 'line', color = 'red')
plt.xlabel('Year', fontsize=12)
plt.ylabel('popularity', fontsize=12)
```

```python
df4 = df.groupby('release_year')['vote_average'].mean()
df4.plot(kind = 'line', color = 'green')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Rating', fontsize=12)
```

```python
df5 = df.plot.scatter(x = 'popularity', y = 'vote_average', c = 'green', figsize=(6,4))
```

```python
split = ['genres']
for i in split:
    df[i] = df[i].apply(lambda x: x.split('|'))
df.head()
```

```python
df = df.explode('genres')
df
```

```python
df['release_date'] = pd.to_datetime(df['release_date'])
df.head()
```

```python
df['extracted_month'] = df['release_date'].dt.month
df.head()
```

```python
df8 = df.groupby('extracted_month')['popularity'].sum()
```

```python
data = {
    'extracted_month': df8.index,
    'popularity': df8.values
}
df8 = pd.DataFrame(data)
```

```python
df8
```

```python
index_to_month = {
    1: 'January',
    2: 'February',
    3: 'March',
    4: 'April',
    5: 'May',
    6: 'June',
    7: 'July',
    8: 'August',
    9: 'September',
    10: 'October',
    11: 'November',
    12: 'December'
}
```

```python
df8.extracted_month = df8.extracted_month.map(index_to_month)
```

```python
df8
```

```python
df8.plot(kind='bar', x='extracted_month', y='popularity', color='darkblue', figsize=(12, 6))
```

```python
df9 = df.groupby('extracted_month')['revenue'].sum()
df9
```

```python
data = {
    'extracted_month': df9.index,
    'revenue': df9.values
}
df9 = pd.DataFrame(data)
```

```python
index_to_month = {
    1: 'January',
    2: 'February',
    3: 'March',
    4: 'April',
    5: 'May',
    6: 'June',
    7: 'July',
    8: 'August',
    9: 'September',
    10: 'October',
    11: 'November',
    12: 'December'
}
```

```python
df9.extracted_month = df9.extracted_month.map(index_to_month)
```

```python
df9.plot(kind='bar', x='extracted_month', y='revenue', color='darkblue', figsize=(12, 6))
plt.title('Monthly revenue of movies', fontsize=16)
plt.xlabel('Revenue', fontsize=12)
plt.ylabel('Month', fontsize=12)
plt.show()
```
